```python
import pandas as pd
import scipy.stats as stats1
from scipy.stats import shapiro
import numpy as np
from scipy.stats import mannwhitneyu
from scipy import stats as stats2
```

```python
champions = pd.read_csv('champions.csv')
stats = pd.read_csv('stats(2).csv')
```

```python
ch_2020 = stats[(stats['season_id'] == 2) & (stats['team_id'] == 26)]
mean_ch_2020 = ch_2020['innate_ability'].mean()
print(mean_ch_2020)
ch_2020
```

```python
ch_2021 = stats[(stats['season_id'] == 3) & (stats['team_id'] == 13)]
mean_ch_2021 = ch_2021['innate_ability'].mean()
print(mean_ch_2021)
ch_2021
```

```python
ch_2022 = stats[(stats['season_id'] == 4) & (stats['team_id'] == 14)]
mean_ch_2022 = ch_2022['innate_ability'].mean()
print(mean_ch_2022)
ch_2022
```

```python
ch_2023 = stats[(stats['season_id'] == 5) & (stats['team_id'] == 15)]
mean_ch_2023 = ch_2023['innate_ability'].mean()
print(mean_ch_2023)
ch_2023
```

```python
group1_data = pd.concat([ch_2020['innate_ability'], ch_2021['innate_ability']])
group2_data = pd.concat([ch_2022['innate_ability'], ch_2023['innate_ability']])
```

```python
#checking if innate_ability data is normal.
stat1, p1 = shapiro(group1_data)
print(f"Statistic: {stat1}, p-value: {p1}")

if p1 > 0.05:
    print("Group1 data is normal.")
else:
    print("Group1 data is not normal.")

#-----------------------------------------------#
stat2, p2 = shapiro(group2_data)
print(f"Statistic: {stat2}, p-value: {p2}")

if p2 > 0.05:
    print("Group2 data is normal.")
else:
    print("Group2 data is not normal.")
```

```python
#We realized that the data is not normal.
#We need to see if we can normalize it or not! For this, we're gonna use Z-score normalization.
```

```python
mean1 = np.mean(group1_data)
std1 = np.std(group1_data)

z_scores1 = (group1_data - mean1) / std1
print(z_scores1)
```

```python
mean2 = np.mean(group2_data)
std2 = np.std(group2_data)

z_scores2 = (group2_data - mean1) / std2
print(z_scores2)
```

```python
#again checking are they normal rn?
stat1, p1 = shapiro(z_scores1)
print(f"Statistic: {stat1}, p-value: {p1}")

if p1 > 0.05:
    print("Group1 data is normal.")
else:
    print("Group1 data is not normal.")
#-----------------------------------------------#
stat2, p2 = shapiro(z_scores2)
print(f"Statistic: {stat2}, p-value: {p2}")

if p2 > 0.05:
    print("Group2 data is normal.")
else:
    print("Group2 data is not normal.")
```

```python
#As we see, data is still not normal, so we must go on plan B, we should use non-parametric methods like Mann-Whitney U!
```

```python
stat, p_value = mannwhitneyu(group1_data, group2_data)

print("U Statistic:", stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("We reject the null hypothesis: The distributions are different.")
else:
    print("We fail to reject the null hypothesis: The distributions are the same.")
```

```python
#So it means that: The average innate ability of players in the champion team, has not increased significantly.
```

## validation of Mann-Whitney U test using the confidence interval method

```python
sample_mean_group1_data, sample_mean_group2_data = np.mean(group1_data), np.mean(group2_data)
sample_std_group1_data, sample_std_group2_data = np.std(group1_data, ddof=1), np.std(group2_data, ddof=1)  # ddof=1 for sample standard deviation
sample_mean_diff = sample_mean_group1_data - sample_mean_group2_data

standard_error_group1_data = sample_std_group1_data / np.sqrt(len(group1_data))
standard_error_group2_data = sample_std_group2_data / np.sqrt(len(group2_data))
standard_error_diff = np.sqrt((sample_std_group1_data**2 / len(group1_data)) + (sample_std_group2_data** 2 / len(group2_data)))
                              
print(f'group1_data  mean: {sample_mean_group1_data:.2f} | group1_data  std: {sample_std_group1_data:.2f}')
print(f'group2_data  mean: {sample_mean_group2_data:.2f} | group2_data  std: {sample_std_group2_data:.2f}')

# Determine the critical value for a 95% confidence interval (two-tailed)
confidence_level = 0.95
df_group1_data = len(group1_data) - 1
df_group2_data = len(group2_data) - 1
df_diff = len(group1_data) - 1 + len(group2_data) - 1

critical_value_group1_data = stats2.t.ppf((1 + confidence_level) / 2, df_group1_data)
critical_value_group2_data = stats2.t.ppf((1 + confidence_level) / 2, df_group2_data)
critical_value_diff = stats2.t.ppf((1 + confidence_level) / 2, df_diff)

# margin of error
margin_of_error_group1_data = critical_value_group1_data * standard_error_group1_data
margin_of_error_group2_data = critical_value_group2_data * standard_error_group2_data
margin_of_error_diff = critical_value_diff * standard_error_diff

# confidence interval
confidence_interval_group1_data = (sample_mean_group1_data - margin_of_error_group1_data, sample_mean_group1_data + margin_of_error_group1_data)
confidence_interval_group2_data = (sample_mean_group2_data - margin_of_error_group2_data, sample_mean_group2_data + margin_of_error_group2_data)
confidence_interval_diff = (sample_mean_diff - margin_of_error_diff, sample_mean_diff + margin_of_error_diff)
lower_lim, upper_lim = confidence_interval_diff

print(f'confidence interval for group1 is {confidence_interval_group1_data}')
print(f'confidence interval for group2 is {confidence_interval_group2_data}')
print(f'confidence interval for different is {confidence_interval_diff}')


if (lower_lim < 0 and upper_lim < 0) or ((lower_lim > 0 and upper_lim > 0)):
    print("Reject the null hypothesis: The average innate ability of players in the champion team, has increased significantly.")
    if lower_lim > 0:
        print("The average innate ability of group1 is higher than group2.")
    else:
        print("The average innate ability of group2 is higher than group1.")
else:
    print("Fail to reject the null hypothesis: The average innate ability of players in the champion team, has not increased significantly.")
```
