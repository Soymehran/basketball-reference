```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('stats(2).csv')
players = pd.read_csv('players.csv')
players = players.rename(columns={'id': 'player_id'})
```

```python
players_ch2022 = (stats[(stats['season_id'] == 4) & (stats['team_id'] == 14)][['player_id' , 'experience']]
                  .sort_values(by = 'experience' , ascending = False)
                 )
merged_height_ch2022 = players_ch2022.merge(players[['player_id' , 'height']], on = 'player_id')
merged_height_ch2022
```

```python
players_ch2023 = (stats[(stats['season_id'] == 5) & (stats['team_id'] == 15)][['player_id' , 'experience']]
                  .sort_values(by = 'experience' , ascending = False)
                 )
merged_height_ch2023 = players_ch2023.merge(players[['player_id' , 'height']], on = 'player_id')
merged_height_ch2023
```

```python
best_15_2022 = (stats[(stats['season_id'] == 4)]
                .sort_values(by = 'Points' ,ascending = False).head(15)
                [['player_id','Points', 'experience']]
               )
merged_height_15p2022 = best_15_2022.merge(players[['player_id' , 'height']], on = 'player_id').sort_values(by = 'experience' , ascending = False)
merged_height_15p2022
```

```python
best_15_2023 = (stats[(stats['season_id'] == 5)]
                .sort_values(by = 'Points' ,ascending = False).head(15)
                [['player_id','Points', 'experience']]
               )
merged_height_15p2023 = best_15_2023.merge(players[['player_id' , 'height']], on = 'player_id').sort_values(by = 'experience' , ascending = False)
merged_height_15p2023
```

```python
sns.histplot(merged_height_15p2022['height'], kde=True, label='Top 15 2022', color='blue', bins=15,  stat = 'density')
sns.histplot(merged_height_ch2022['height'], kde=True, label='Champion 2022', color='red', bins=15,  stat = 'density')
plt.legend()
plt.title("Height Distribution: Champion vs Top 15 (2022)")
plt.show()
```

```python
sns.histplot(merged_height_15p2023['height'], alpha = 0.6, kde=True, label='Top 15 2023', color='black', bins=12, stat = 'density')
sns.histplot(merged_height_ch2023['height'],alpha = 0.6, kde=True, label='Champion 2023', color='blue', bins=12,  stat = 'density')
plt.legend()
plt.title("Height Distribution: Champion vs Top 15 (2023)")
plt.show()
####   champions
```

```python
sns.histplot(merged_height_ch2022['experience'], kde=True, label='Champion 2022', color='purple', bins=15, stat ='density')
sns.histplot(merged_height_15p2022['experience'], kde=True, label='Top 15 2022', color='cyan', bins=15, stat ='density')
plt.legend()
plt.title("Experience Distribution: Champion vs Top 15 (2022)")
plt.show()
```

```python
sns.histplot(merged_height_ch2023['experience'], kde=True, label='Champion 2023', color='green', bins=12, stat = 'density')
sns.histplot(merged_height_15p2023['experience'], kde=True, label='Top 15 2023', color='orange', bins=12, stat = 'density')
plt.legend()
plt.title("Experience Distribution: Champion vs Top 15 (2023)")
plt.show()
```
